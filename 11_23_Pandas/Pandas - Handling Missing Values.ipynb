# Pandas - Handling Missing Values

```python
import pandas as pd
import numpy as np
```

## Pandas Functions

```python
1. Detect Missing Values:
    1. df.isna()
    2. df.isnull()
    3. df.isna().sum()
    4. df.isnull().sum()
    5. df.isna().mean()
    6. df.isna().mean()*100
    7. df['Col_name'].isna().any() >>  True or False 
    
2. Drop Null Values:
    1. df.dropna(thresh = col_count, axis=0, inplace=False)
       df.dropna(thresh= row_count, axis=1, inplace = False)
       df.dropna(subset =['Col_Name']) >> Delete rows in terms of column
    
    2. df.drop('Col_Name', axis = 1, inplace=False)
    3. df.drop([Col1, Col2,....ColN], axis=1, inplace=False)
    4. df.drop('Row NAme', inplace=False)
       df.drop('Row NAme', axis=0, inplace=False)
        
    5. df.drop([Row1,Row2,...RowN], axis = 0, inplace=False)
    
3. Impute Missing Values:
    1. df.fillna(value, inplace= False)
    
    Statistical Methods:
        1. df['Col_name'].mean()
        2. df['Col_name'].median()
        3. df['Col_name'].mode()[0]
        
Mean >> If outliers are not present in data
Median >> If outliers are present in data
Mode  >> Categorical Data
```

```python
Missing Values are represented by NaN

NaN >> Null Values >> Not a Number

Datatype of NaN >> float
```
